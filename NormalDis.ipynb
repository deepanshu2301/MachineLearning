```python
import numpy as np
```

```python
m1=np.array([7,8])
v1=np.array([ [1,0],[0,2] ])
m2=np.array([1,1])
v2=np.array([ [1,0.3],[0.3,1] ])

z1=np.random.multivariate_normal(m1,v1,500)
z2=np.random.multivariate_normal(m2,v2,500)
plt.scatter(z1[:,0],z1[:,1])
plt.scatter(z2[:,0],z2[:,1])
plt.show()
```
