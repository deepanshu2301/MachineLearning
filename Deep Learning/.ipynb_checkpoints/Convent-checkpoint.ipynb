```python
from keras.layers import *
from keras.models import Sequential
```

```python
## Building a model

model = Sequential()
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D((2,2)))

model.add(Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(MaxPool2D((2,2)))

model.add(Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(Flatten())
model.add(Dense(64,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
from keras.datasets import mnist
from keras.utils import to_categorical
```

```python
(XTrain,YTrain),(XTest,YTest) = mnist.load_data()
```

```python
def preprocess_data(X,Y):
    X=X.reshape((-1,28,28,1))
    X=X/255.0
    Y=to_categorical(Y)
    return X,Y
```

```python
XTrain,YTrain = preprocess_data(XTrain,YTrain)
print(XTrain.shape,YTrain.shape)
```

```python
XTest,YTest = preprocess_data(XTest,YTest)
print(XTest.shape,YTest.shape)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
hist=model.fit(XTrain,YTrain,epochs=10,validation_split=0.1,batch_size=128)
```

```python
model.evaluate(XTest,YTest)
```

```python
import matplotlib.pyplot as plt
plt.figure(0)
plt.plot(hist.history['loss'],'g')
plt.plot(hist.history['val_loss'],'b')
plt.plot(hist.history['acc'],'r')
plt.plot(hist.history['val_acc'],'black')
plt.show()
```

## data augmentation (scaling,resizing,rotating the same image to expand our dataset)

```python
img_path = "E:\Python Folder\Datasets\Images\dogs\dog.8.jpg"
```

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
img = image.load_img(img_path,target_size=(100,100))
```

```python
x=image.img_to_array(img)/255.0
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```

```python
x=x.reshape((1,100,100,3))
```

```python
datagen = image.ImageDataGenerator(
rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)
```

```python
i=0
for batch in datagen.flow(x,batch_size=1):
    plt.figure()
    imgplot = plt.imshow(image.img_to_array(batch[0]))
    i+=1
    if i==10:
        break
    plt.show()
```
