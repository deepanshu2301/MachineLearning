```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.randn(500)
u=0
std=1
x=std*x +u
noise=np.random.randn(500)
y=3*x+2+noise
plt.scatter(x,y)
plt.show()
```

```python
u1=np.array([2,4])
cor1=np.array([[2,0.8],[0.8,1]])
u2=np.array([3,1])
cor2=np.array([[5,-2],[-2,1]])

x1=np.random.multivariate_normal(u1,cor1,300)
x2=np.random.multivariate_normal(u2,cor2,400)
plt.scatter(x1[:,0],x1[:,1],color='green')
#plt.scatter(x2[:,0],x2[:,1],color='red')

y=np.zeros(700)
y[:300]=1
x=np.vstack((x1,x2))
plt.show()
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```
