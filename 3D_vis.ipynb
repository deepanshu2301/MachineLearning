```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=[-3,-2,-1,0,1,2,3]
b=a
a,b=np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
c=np.linspace(-4,4)
d=c
d,c=np.meshgrid(c,d)
print(c)
print(d)
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(c,d,c**2+d**2,cmap='rainbow')
plt.show()
```
