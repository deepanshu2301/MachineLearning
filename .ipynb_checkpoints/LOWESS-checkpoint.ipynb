LOGICALLY WEIGHTED REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv("Datasets\Locally Weighted Regression\WeightedX.csv")
dfy=pd.read_csv("Datasets\Locally Weighted Regression\WeightedY.csv")
```

```python
X=dfx.values
Y=dfy.values
print(X.shape)
print(Y.shape)
```

```python
#Normalize the Data
```

```python
X=(X-X.mean())/X.std()
plt.scatter(X,Y)
plt.show()
```

```python
def getW(querypt,X,tau):
    M=X.shape[0]
    W=np.mat(np.eye(M))
    
    for i in range(M):
        xi=X[i]
        x=querypt
        W[i,i]=np.exp(np.dot(xi-x,(xi-x).T)/(-2*tau*tau))
    
    return W
        
        
        
```

```python
X=np.mat(X)
Y=np.mat(Y)
M=X.shape[0]

W=getW(-1,X,0.5)
print(W.shape)
print(W)
```

```python
def predict(X,Y,querypt,tau):
    ones=np.ones((M,1))
    X_=np.hstack((X,ones))
    
    qx=np.mat([querypt,1])
    W=getW(qx,X_,tau)
    
    theta=np.linalg.pinv(X_.T*(W*X_))*(X_.T*(W*Y))
    
    pred=np.dot(qx,theta)
    return theta,pred
    
```

```python
theta,pred=predict(X,Y,1.0,1.0)
print(theta)
print(pred)
```

VISUALIZE 

```python
def plotprediction(tau):
    X_test=np.linspace(-2,2,20)
    Y_test=[]
    
    for xq in X_test:
        theta,pred=predict(X,Y,xq,tau)
        Y_test.append(pred[0][0])
        
    Y_test=np.array(Y_test)
    
    XO=np.array(X)
    YO=np.array(Y)
    
    plt.scatter(XO,YO)
    plt.scatter(X_test,Y_test,color='red')
    plt.show()
```

```python
plotprediction(1)
```

```python
taus=[0.1,0.5,1.0,10]
for i in taus:
    plotprediction(i)
```
