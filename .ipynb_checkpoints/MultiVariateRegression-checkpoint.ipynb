```python
def hypothesis(theta,x):
    return theta[0] + theta[1]*x[0] + theta[2]*x[1]+ theta[3]*x[2]+ theta[4]*x[3]+ theta[5]*x[4]

def error(X,Y,theta):
    m=X.shape[0]
    error=0
    for i in range(m):
        hp=hypothesis(theta,X[i])
        error+=(hp-Y[i])**2
    return error

def gradient(X,Y,theta):
    
    grad=np.zeros((6,))
    m=X.shape[0]
    
    for i in range(m):
        hp=hypothesis(theta,X[i])
        grad[0]+=(hp-Y[i])
        grad[1]+=(hp-Y[i])*X[i,0]
        grad[2]+=(hp-Y[i])*X[i,1]
        grad[3]+=(hp-Y[i])*X[i,2]
        grad[4]+=(hp-Y[i])*X[i,3]
        grad[5]+=(hp-Y[i])*X[i,4]
        
    return grad

def gradientdesc(X,Y,n=0.001):
    
    theta=np.zeros((6,))
    itr=0
    mitr=100
    errorlist=[]
    thetalist=[]
    while(itr<=mitr):
        grad=gradient(X,Y,theta)
        errorx=error(X,Y,theta)
        errorlist.append(errorx)
        
        thetalist.append((theta[0],theta[1],theta[2],theta[3],theta[4],theta[5]))
        theta[0]= theta[0] - n*grad[0]
        theta[1]= theta[1] - n*grad[1]
        theta[2]= theta[2] - n*grad[2]
        theta[3]= theta[3] - n*grad[3]
        theta[4]= theta[4] - n*grad[4]
        theta[5]= theta[5] - n*grad[5]
        
        itr+=1
    return theta,errorlist,thetalist

```

```python
finaltheta1,final_err1,theta_list=gradientdesc(X,Y)
print(finaltheta1)
```

```python
plt.plot(final_err1)
plt.show()
```
